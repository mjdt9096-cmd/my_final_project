```python
import pandas as pd
import os
from google.colab import files

WORK_DIR = '/content/prompt_files'
os.makedirs(WORK_DIR, exist_ok=True)

def upload_and_rename(prompt_num):
    print(f"ارفع ملف البرومبت رقم {prompt_num}")
    uploaded = files.upload()

    if len(uploaded) != 1:
        print(f"يجب رفع ملف واحد فقط، تم رفع {len(uploaded)}")
        return None

    original_name = list(uploaded.keys())[0]
    df = pd.read_csv(original_name)

    file_prompt_ids = df['prompt_id'].unique()
    print(f"\nالملف: {original_name}")
    print(f"  عدد المقالات: {df.shape[0]}, عدد الأعمدة: {df.shape[1]}")
    print(f"  prompt_id داخل الملف: {file_prompt_ids}")

    if len(file_prompt_ids) > 1:
        print(f"  تنبيه: الملف يحتوي على أكثر من برومبت")
    elif file_prompt_ids[0] != prompt_num:
        print(f"  خطأ: المتوقع برومبت {prompt_num} لكن الملف يحتوي برومبت {file_prompt_ids[0]}")

    new_name = f'{WORK_DIR}/prompt_{prompt_num}.csv'
    df.to_csv(new_name, index=False)
    if os.path.exists(original_name):
        os.remove(original_name)
    print(f"  تم الحفظ كـ: prompt_{prompt_num}.csv")
    return new_name

print("تم تجهيز الإعدادات")
```

```python
import os
import pandas as pd

prompt_files = sorted([f for f in os.listdir(WORK_DIR) if f.endswith('.csv')])
print(f"عدد الملفات الموجودة: {len(prompt_files)}\n")

summary = []
for f in prompt_files:
    df = pd.read_csv(f'{WORK_DIR}/{f}')
    summary.append({
        'file': f,
        'essays': df.shape[0],
        'cols': df.shape[1],
        'prompt_ids': sorted(df['prompt_id'].unique()),
        'nulls': df.isnull().sum().sum()
    })

summary_df = pd.DataFrame(summary)
print(summary_df.to_string(index=False))

all_prompts = []
for s in summary:
    all_prompts.extend(s['prompt_ids'])

print(f"\nمجموع البرومبتات: {sorted(all_prompts)}")

if len(all_prompts) != len(set(all_prompts)):
    print("تنبيه: يوجد برومبت مكرر")
elif len(set(all_prompts)) < 8:
    missing = set(range(1, 9)) - set(all_prompts)
    print(f"ناقص برومبتات: {sorted(missing)}")
else:
    print("كل البرومبتات الثمانية موجودة وغير مكررة")
```

```python
import pandas as pd
import os

prompt_files = sorted([f for f in os.listdir(WORK_DIR) if f.endswith('.csv')])

all_dfs = []
print("الملفات التي ستدمج:\n")
for f in prompt_files:
    df = pd.read_csv(f'{WORK_DIR}/{f}')
    all_dfs.append(df)
    print(f"  {f}: {df.shape[0]} مقالة × {df.shape[1]} عمود")

merged = pd.concat(all_dfs, ignore_index=True, sort=False)
print(f"\nالملف الموحد قبل المعالجة: {merged.shape[0]} مقالة × {merged.shape[1]} عمود")

# الأعمدة الموجودة في بعض البرومبتات فقط تصير null - نملأها بصفر
META_COLS = ['essay_id', 'prompt_id', 'essay', 'prompt', 'prompt_type', 'prompt_type_encoded',
             'relevance', 'organization', 'vocabulary', 'style', 'development',
             'mechanics', 'grammar', 'holistic']

null_count_before = merged.isnull().sum().sum()
feature_cols = [c for c in merged.columns if c not in META_COLS]
merged[feature_cols] = merged[feature_cols].fillna(0)

print(f"تم ملء {null_count_before} قيمة فارغة بصفر")

print(f"\nالملف النهائي:")
print(f"  المقالات: {merged.shape[0]}")
print(f"  الفيتشرز: {merged.shape[1]}")
print(f"\nتوزيع المقالات على البرومبتات:")
print(merged['prompt_id'].value_counts().sort_index().to_string())
```

```python
from google.colab import files

output_path = 'laila_features_all_prompts.csv'
merged.to_csv(output_path, index=False)
print(f"تم حفظ الملف: {output_path}")
print(f"الأبعاد: {merged.shape[0]} × {merged.shape[1]}")
print(f"\nجاري التنزيل...")
files.download(output_path)
```
